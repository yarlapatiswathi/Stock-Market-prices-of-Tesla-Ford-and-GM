### Market capitalization of Tesla, Ford and GM

+ What is market capitalization?

- It refers to the total value of all a company's shares of stock. 
- It is calculated by multiplying the $price_of_a_stock$ by its $total_number_of_outstanding_shares$.
- For example, a company with 20 million shares selling at 50 dollars a share would have a market cap of 1 billion dollars.
- Estimated Market cap = volume * open stock price

```python
tesla['Total Traded']= tesla['Open']*tesla['Volume']
ford['Total Traded']=ford['Open']*ford['Volume']
gm['Total Traded']=gm['Open']*gm['Volume']
```

```python
tesla['Total Traded'].plot(label='Tesla',figsize=(15,7))
ford['Total Traded'].plot(label='Ford')
gm['Total Traded'].plot(label='GM')
plt.legend()
plt.ylabel('Total Traded')
plt.title('Market Cap of Tesla, Ford and GM')
plt.show()
```

- There seems to be a huge amount of money traded for tesla stocks in the end of 2020.
- we can see that there is lot of noise in the plot.
- Also when compared to Tesla, We cannot clearly observe the trading value of Ford and GM.

### Moving Average(Rolling mean):

- A method of smoothening the curve by calculating the average of data over period of time.

```python
tesla['MA50']=tesla['Open'].rolling(50).mean()
ford['MA50']=ford['Open'].rolling(50).mean()
gm['MA50']=gm['Open'].rolling(50).mean()
```

```python
tesla['MA50'].plot(label='Tesla',figsize=(15,7))
plt.legend()
plt.show()
```

- Tesla has see a huge spike in stock price in 2021 and kept growing and is in fall in 2023.

```python
ford['MA50'].plot(label='Ford',figsize=(15,7))
```

- There was a huge spike in Ford's stock price in 2021 and kept falling down till now.

```python
gm['MA50'].plot(label='GM',figsize=(15,7))
```

- GM and Ford are on almost on the same page. Even GM has seen a huge spike in 2021 and again in 2022 but kept falling down after that. Finally the stock price is seeing much spike in the stock price.
- We can see the trends by using moving average as it reduces the noise. If rolling window is high we might miss the trends. So it is better to use small rolling window.

### Correlation & Scatter Matrix:

- Correlation Matrix helps to see the relation between the companies and their performance in relation to each other.

```python
car_comp=pd.concat([tesla['Open'],ford['Open'],gm['Open']],axis=1)
car_comp.columns=['Tesla Open','Ford Open','GM Open']
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(car_comp,figsize=(8,8),hist_kwds={'bins':50})
```
